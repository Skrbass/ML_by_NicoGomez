#### Ejemplo:
calcule la fuerza gravitacional entre 2 cuerpos:

````
print("### Calculo de la fuerza gravitacional entre dos cuerpos ###")
m1 = float(input("ingrese la masa 1 en kg: "))
m2 = float(input("ingrese la masa 2 en kg: "))
r = float(input("ingrese la distancia entre ambos cuerpos en metros: "))
G = 6.674E-11 # Constante gravitacional en unidades del sistema internacional (N m**2 / kg**2)
grav = G * ((m1 * m2) / r**2) # Calculo de la fuerza gravitacional
print("la fuerza de atraccion gravitacional entre ambos cuerpos es de: {:.5e} Newtons".format(grav))
````

#### if ... else
- Estructura: 

``` python

if (condicion):

    instrucción

```

se debe dar una condición lógica a partir de operaciónes Booleanas o metodos que retornen  `True` o `False`

#### Ejemplo:

determinar si un número ingresado es par o impar:

````
num = int(input("ingrese un numero:"))
cond=num%2
if (cond==0):
    print("el numero es par")
else:
    print("el numero es impar ")
````

#### Ejemplo 

aproxime pi usando la sumatoria de Gregory: $$\pi = 4\sum_{k=1}^{N}\frac{(-1)^{k+1}}{2k-1}$$ para un numero de iteraciones N dado por el usuario:

````
N = int(input("numero de repeticiones: "))

k = 1.  #contador
pi = 0. 

while k <= N:
  pi += ((-1)**(k + 1))/(2 * k - 1)
  k += 1
pi *= 4

print("pi: %.10f"%(pi))
````

#### Ejemplo

determine si un numero dado es primo:

````
n = int(input("digite un numero natural: "))

sw = 0

if n <= 1:
  print(n," no es primo")
else:
  for i in range(2,n):

    if n%i == 0:

      sw = 1
      break

  if sw == 1:
    print(n," no es primo")
  else:
    print(n," es primo")
````

#### Ejemplo:

Factorial de un número

````
def factorial(x):
    f=1
    for i in range(2,x+1):
        f*=i
    return f

x=int(input("digite un numero: "))

factorial(x)
````

#### Ejemplo:

Derivada numérica por diferencia central:

$$f^\prime (x_0) \approx \frac {f(x_0+h)-f(x_0-h)} {2h}$$

````
def derivada(x, h, f):
  return (f(x + h) - f(x - h)) / (2 * h)
````

````
from math import sin, cos, pi # Importando funciones, más adelante lo veremos con detalle
print(derivada(0, 0.01, sin)) # La derivada de seno es coseno
print(derivada(pi / 2, 0.01, cos)) # La derivada de seno es coseno
````

### Funciones Lambda

Cuando una función es sencilla de construir es buena idea usar funciones Lambda, permitiendo construir funciones en un línea de código. Como se verá más adeante, otra característica importante de las funciones Lambda es que permiten construir funciones de manera rápida sin necesidad de nombrarlas.

``` python

lambda parametros : intrucción

```

#### Ejemplo:

````
duplicar = lambda x : x * 2
````

````
print(duplicar(5))
````

````
suma = lambda x, y : x + y
````

````
suma(2,3)
````
